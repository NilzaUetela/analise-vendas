```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
caminho = "/content/drive/MyDrive/analysis.ipynb/data/vendas.csv"
df = pd.read_csv(caminho, parse_dates=["Data"])
df.head()
```

```python
produtos = df.groupby("Produto")["Valor"].sum().sort_values(ascending=False)
produtos.plot(kind="bar", title="Total de vendas por produto", ylabel="Valor R$", color="skyblue")
plt.show()
```

```python
regioes = df.groupby("Região")["Valor"].sum().sort_values(ascending=False)
regioes.plot(kind="bar", title="Total de vendas por região", ylabel="Valor R$", color="lightgreen")
plt.show()
```

```python
df["Mes"] = df["Data"].dt.to_period("M")
mensal = df.groupby("Mes")["Valor"].sum()
mensal.plot(kind="line", marker="o", title="Evolução mensal das vendas", ylabel="Valor R$")
plt.show()
```

## 📌 Conclusões

- A **Camisa** é o produto mais vendido.
- A **Região Sul** lidera nas vendas.
- O mês de **Março** teve o maior volume de vendas.
